```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
stocks = pd.read_csv('VTI_stocks.csv')
stocks.drop(columns=['Unnamed: 0'], inplace=True)
stocks.head()
```

```python
stocks['Date'] = pd.to_datetime(stocks['Date'])
```

```python
stocks = stocks.set_index('Date')
```

```python
stocks.head()
```

```python
# No non-numeric data
# Scaling
scaler = MinMaxScaler() 
s_stocks = scaler.fit_transform(stocks)
s_stocks = pd.DataFrame(stocks, columns = stocks.columns)
```

```python
s_stocks.head() # looks like no need for normalization
```

```python
# Define data and target
X = s_stocks.drop(columns=['Volume'])
y = s_stocks['Volume']
```

```python
# Save
X.to_csv('X_stocks_vol.csv')
y.to_csv('y_stocks_vol.csv')
```
